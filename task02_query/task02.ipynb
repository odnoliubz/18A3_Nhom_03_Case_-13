```python
import pandas as pd
df_patient = pd.read_csv("../data_clean/patient_info_clean.csv")
df_reaction = pd.read_csv("../data_clean/reaction_log_clean.csv")
df_vaccine = pd.read_csv("../data_clean/vaccination_record_clean.csv")
```

#1: ĐẾM SỐ BỆNH NHÂN THEO NHÓM TUỔI

```python
patient=df_patient['age_group'].value_counts().reset_index(name='số bệnh nhân')
patient
```

#2: TÌM BỆNH NHÂN ĐÃ TIÊM TỪ 2 LOẠI VẮC XIN TRỞ LÊN

```python
vacccine=df_vaccine.groupby("patient_id")["vaccine_type"].nunique().reset_index(name="so_loai_vaccine").query("so_loai_vaccine >= 2")
vacccine
```

#3: THỐNG KÊ SỐ LẦN TIÊM THEO LOẠI VẮC XIN

```python
so_lan_vaccine=df_vaccine.groupby('vaccine_type').size().reset_index(name=('số lần tiêm'))
so_lan_vaccine
```

```python
so_lan_vaccine.plot(kind='bar',x='vaccine_type',y='số lần tiêm',title='SỐ LẦN TIÊM THEO LOẠI VACCINE',color='darkblue',legend=True)
```
