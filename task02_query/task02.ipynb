NHIỆM VỤ 2: TRUY VẤN VÀ THỐNG KÊ MÔ TẢ

```python
import pandas as pd
df_patient = pd.read_csv("../data_clean/patient_info_clean.csv")
df_reaction = pd.read_csv("../data_clean/reaction_log_clean.csv")
df_vaccine = pd.read_csv("../data_clean/vaccination_record_clean.csv")
```

```python
#1: ĐẾM SỐ BỆNH NHÂN THEO NHÓM TUỔI
df_patient['age_group'].value_counts().reset_index(name='số bệnh nhân')
```

```python
#2: TÌM BỆNH NHÂN ĐÃ TIÊM TỪ 2 LOẠI VẮC XIN TRỞ LÊN
df_vaccine.groupby("patient_id")["vaccine_type"].nunique().reset_index(name="so_loai_vaccine").query("so_loai_vaccine >= 2")
```

```python
#3: THỐNG KÊ SỐ LẦN TIÊM THEO LOẠI VẮC XIN
df_vaccine.groupby('vaccine_type').size().reset_index(name=('số lần tiêm theo loại vaccine'))
```

```python
#4: LIỆT KÊ CÁC BẢN GHI CÓ DOSE BẤT THƯỜNG
df_vaccine[df_vaccine['dose_number'].isna()]
```
